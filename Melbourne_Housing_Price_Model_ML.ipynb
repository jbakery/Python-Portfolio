```python
# Import Libraries.

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
```

```python
# Insert Data into Data Frame and print first 5 rows.

df = pd.read_csv('Melbourne_housing_FULL.csv')

df.head()
```

```python
# Creating list of DataFrame columns to understand the features.

df.columns
```

```python
# Printing list of data type associated with each feature.

print(df.dtypes)
```

```python
# Scrubbing Data set, deleting unneeded columns.

del df ['Address']
del df ['Method']
del df ['SellerG']
del df ['Date']
del df ['Postcode']
del df ['Lattitude']
del df ['Longtitude']
del df ['Regionname']
del df ['Propertycount']
```

```python
# Double checking proper columns were deleted.

df.head()
```

```python
# Scrubbing Rows with missing values.

df.dropna(axis = 0, how = 'any', subset = None, inplace = True)
```

```python
# Check for NaN values in the DataFrame.
nan_values = df.isna()

if nan_values.any().any():
    # If there are NaN values, print the DataFrame of NaN values.
    print(nan_values)
else:
    # If there are no NaN values, print a message.
    print("No NaN values")
```

```python
# Convert column with non-numeric data into numeric data with one hot encoding.

df = pd.get_dummies(df, columns= ['Suburb','CouncilArea','Type'])
```

```python
# Checking one hot encoding.

df.head()
```

```python
# Assign sependent variable and independent variable.

x = df.drop('Price', axis = 1)

y = df['Price']
```

```python
# Split the Dataset 70/30 split.

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, shuffle = True)
```

```python
# Select the algorithm and Configure Hyperparameters

model = ensemble.GradientBoostingRegressor (

n_estimators = 250,
learning_rate = 0.1,
max_depth = 5,
min_samples_split = 4,
min_samples_leaf= 6,
max_features = 0.6,
loss = 'huber'

)
```

```python
# Fit the model to the test data set.

model.fit(x_train, y_train)
```

```python
# Evaluate model results with test Data.

mae_train = mean_absolute_error( y_train, model.predict(x_train))

print("Training Set Mean Absolute Error: %0.f" % mae_train)

mae_test = mean_absolute_error( y_test, model.predict(x_test))

print("Test Set Mean Absolute Error: %0.f" % mae_test)
```
