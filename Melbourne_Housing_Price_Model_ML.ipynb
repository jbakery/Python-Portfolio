```python
# Scrubbing Data set, deleting unneeded columns.

del df ['Address']
del df ['Method']
del df ['SellerG']
del df ['Date']
del df ['Postcode']
del df ['Lattitude']
del df ['Longtitude']
del df ['Regionname']
del df ['Propertycount']
```
