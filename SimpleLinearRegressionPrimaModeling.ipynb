```python
#Importing Python Packages for Simple Linear Regression

from pydataset import data
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
# %matplotlib notebook will lead to interactive plots embedded within the notebook.

%matplotlib inline
```

```python
#We will be using the Pima women Data set this is about Native american Women living in Pheonix. 
#What we're going to attemp to do is use triceps skinfold measurements to predic BMI.

pima = data('Pima.tr')
```

```python
#first what we want, to see if this data looks approximately linear we plot BMI on the y axis 
#and skin = skinfold measurements on the x - axis.

pima.plot( kind = 'scatter', x = 'skin', y = 'bmi', color = 'b')
plt.title('Prima Data Scatter Plot')
plt.figure
plt.show
```

```python
#That looks pretty linear so we are going to proceed with our linear regression model.
#Now we are going to do a test train split. We are doing supervised learning.

x_train, x_test, y_train, y_test = train_test_split(pima.skin, pima.bmi)
```

```python
#Test train split visualization
plt.figure()
plt.scatter(x_train, y_train, label = 'Training Data', color = 'r', alpha = .7)
plt.scatter(x_test, y_test, label = 'Testing Data', color='g', alpha = .7)
plt.legend()
plt.title('Test Train Split')
plt.show()
```

```python
# Create linear model and train it
LR = LinearRegression()
LR.fit(x_train.values.reshape(-1,1), y_train.values)
```

```python
# Use model to predict on Test data
prediction = LR.predict(x_test.values.reshape(-1,1))

# Plot prediction line against actual test data
plt.figure()
plt.plot(x_test, prediction, label = 'Linear Regression', color ='b')
plt.scatter(x_test, y_test, label = 'Actual Test Data', color = 'g', alpha =.7)
plt.title('Prediction Plot')
plt.legend()
plt.show()
```

```python
# Prediction BMI of Women With Skin fold 50
LR.predict(np.array([[50]]))[0]
```

```python
#Score this model R value
LR.score(x_test.values.reshape(-1,1),y_test.values)
```
